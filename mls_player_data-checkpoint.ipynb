```python
query = """
SELECT position, 
    MIN(born) as min_position_age
FROM mls_player_data
GROUP BY position
ORDER BY MIN(born) DESC;
""" 

df1 = pd.read_sql(query, conn)
df1.head(20)
```

```python
query = """
SELECT full_name, 
    nation,
    goals,
    matches_played,
    1.0 * goals/matches_played as goals_per_game
FROM mls_player_data
WHERE nation = 'USA';
""" 

df1 = pd.read_sql(query, conn)
df1.head(20)
```

```python
query = """
SELECT full_name,
    nation,
    CASE 
        WHEN nation = 'USA' THEN 'NA' 
        WHEN nation = 'ENG' THEN 'EU'
        WHEN nation IN ('URU', 'HON') THEN 'SA'
        ELSE 'AF' 
    END as player_continent
FROM mls_player_data
ORDER BY 3;
""" 
# order by 3 is talking about columns in what we're selecting

df1 = pd.read_sql(query, conn)
df1.head(20)
```

```python
query = """
SELECT SUM(goals) AS total_goals,
    SUM(CASE WHEN nation = 'USA' THEN goals ELSE 0 END) AS goals_USA,
    1.0 * SUM(CASE WHEN nation = 'USA' THEN goals ELSE 0 END) / SUM(goals) AS percentage_USA_goals
FROM mls_player_data;
""" 

df1 = pd.read_sql(query, conn)
df1.head(20)
```

```python
query = """
SELECT SUM(assists) AS total_assists,
    SUM(CASE WHEN born > 1994 THEN assists ELSE 0 END) AS assists_after_1994,
    1.0 * SUM(CASE WHEN born > 1994 THEN assists ELSE 0 END) / SUM(assists) AS pcnt_assists_1994
FROM mls_player_data;
""" 

df1 = pd.read_sql(query, conn)
df1.head(20)
```
